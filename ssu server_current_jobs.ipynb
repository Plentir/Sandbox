```python
import pandas as pd
```

```python
raw = open("Current_running_jobs.txt", "r", encoding="utf-8")

lines = []
lines = raw.readlines()

print(*lines[:10], sep="\n")
```

```python
lines_proc = []

for l in lines:
    pre = l.split()
    words = [pre[0], pre[1], " ".join(pre[2:])]
    lines_proc.append(words)
```

```python
import pprint
pprint.pprint(lines_proc[:10])
```

```python
# * (Splat Operator)를 활용하면 리스트를 줄 바꿈하여 출력 가능!
print(*lines_proc[:10], sep="\n")
```

```python
(col_1, col_2, col_3) = (lines_proc[0][0],
                         lines_proc[0][1],
                         lines_proc[0][-1])

data = pd.DataFrame(lines_proc[1:]).rename(columns={0: col_1,
                                                    1: col_2,
                                                    2: col_3})

data
```

```python
data = data.sort_values(by=["UID", "PID"])
data
```

```python
usr_list = set(list(data["UID"]))
trg_list = sorted([i for i in usr_list if "bee" in i])

res = {}
for target in trg_list:
    res[target] = (data.loc[data["UID"] == target, :])

print(len(res))
print(*res.keys())
```

# Markdown Basic
------
1. Block
    - Code
        + Inline  
            Use \` like this, `x, y = 1, 2`.
        + Code Fence  
            ```
            x, y = 3, 4
            print(x, y)
            ```
    - Quote
        + Indent block  
            > Use indent (4 spaces.) to make quote blocks.            
        + Quote block  
            > Use \> to make quote blocks.  
            > Continuous \> are making one quote block.
            >> You can make nested one with series of \>.
            
            
2. Text type
    - Bold  
        Use two * (which means, \*\*) to make a **bold**.
    - Italic  
        Use one * (which means, \*) to make an *italic*.
    - Deleted  
        Use ~ to make a line ~middle of words~.
    - Combination  
        You ***can*** ~*combine* **those**~.


3. Text structure
    - Break line  
        Use \\ at the end of line to break\
        line.  
    - Paragraph  
        Insert two line breaks (Enter \* 2).
        
        Then their paragraphs would be separated.
        
        
4. Link & Images
    - Link  
        [Text to show as link](Address to link to text)    
        [Like this](#Markdown-Basic)
    - Images  
        ![Alternative message](Address of image)  
        ![Example](https://assets.ubuntu.com/v1/8dd99b80-ubuntu-logo14.png)

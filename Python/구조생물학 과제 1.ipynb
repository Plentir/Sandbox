```python
def print_seq(string1, string2, charPerCol=10, maxCol=5):
    cells = int(len(string1)/charPerCol)
    
    def opr(string, cells, charPerCol):
        chars = list(string.replace(" ", "-"))
        res = []
        
        for i in range(cells):
            res.append("".join(chars[:charPerCol]))
            del chars[:charPerCol]
            
        res.append("".join(chars[0:]))
        
        return res

    res_s1 = opr(string1, cells, charPerCol)
    res_s2 = opr(string2, cells, charPerCol)
    
    # [res_s1, res_s2, current_cols]
    cursor = [0, 0, 0]
    
    while cursor[1] < (cells - 3):
        print("\n")
        while (cursor[-1] < (maxCol)):
            print(res_s1[cursor[0]], end="\t")
            cursor[0] += 1
            cursor[-1] += 1
        
        print("")
        for col in range(cursor[-1]):
            print(res_s2[cursor[1]], end="\t")
            cursor[1] += 1
        cursor[-1] = 0
        
    print("\n")
    print(*res_s1[cursor[0]:], sep="\t")
    print(*res_s2[cursor[1]:], sep="\t")
        
    return
```

```python
def reformatter(ss, charPerCol=10, maxCol=5):
    res = ""
    
    for c in ss:
        if c in ["G", "H", "I"]:
            res += "H"
        
        elif c == "E":
            res += "E"
            
        else:
            res += "C"
            
    
    if len(ss) == len(res):
        print("Jobs done.", end="")
        
    else:
        print("Jobs canceled because of errors.")
        return 0
    
    print_seq(ss, res, charPerCol, maxCol)
    return res
```
